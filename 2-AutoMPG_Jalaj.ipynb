```python
import pandas as pd
```

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(path, names=headers, na_values=["?"])
df.head(5)
```

```python
import numpy as np

outliers = []

for feature in df.select_dtypes(include=np.number).columns:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    if df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].any(axis=None):
        outliers.append(feature)

print("Attributes with outliers:", outliers)
```

```python
# Handling Missing Values
# 1. Replace missing values in 'normalized-losses' (41 missing) with the mean, since it's a significant proportion of the dataset.
df["normalized-losses"] = df["normalized-losses"].fillna(df["normalized-losses"].mean())

# 2. Replace missing values in 'num-of-doors' (2 missing) with the mode, as it's a categorical variable and only 2 missing values.
df["num-of-doors"] = df["num-of-doors"].fillna(df["num-of-doors"].mode()[0])

# 3. Replace missing values in 'bore' (4 missing) using backward fill, since it may have an increasing trend or sequence.
df["bore"] = df["bore"].bfill()

# 4. Replace missing values in 'stroke' (4 missing) using the mean, assuming it's normally distributed and the missing proportion is small.
df["stroke"] = df["stroke"].fillna(df["stroke"].mean())

# 5. Replace missing values in 'horsepower' (2 missing) using the median, as the data might be skewed.
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())

# 6. Replace missing values in 'peak-rpm' (2 missing) using the mean, since the missing proportion is very small.
df["peak-rpm"] = df["peak-rpm"].fillna(df["peak-rpm"].mean())

# 7. Replace missing values in 'price' (4 missing) using the median, as it's likely skewed and is a crucial variable.
df["price"] = df["price"].fillna(df["price"].median())

# Step 3: Verifying that no missing values remain
print("Missing values after handling:\n", df.isnull().sum())
```
