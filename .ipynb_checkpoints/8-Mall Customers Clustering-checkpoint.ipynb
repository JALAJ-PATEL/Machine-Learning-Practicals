# Mall Customers Clustering

# Problem Statement

You own a supermarket mall and, through membership cards, have collected basic data about your customers, including:

- **Customer ID:** A unique identifier for each customer.
- **Age:** The age of the customer.
- **Gender:** The gender of the customer (Male/Female).
- **Annual Income:** The annual income of the customer (in $1000s).
- **Spending Score:** A score assigned to the customer based on their behavior and purchasing patterns.

## Objective:
As the mall owner, you want to understand your customers to identify target customer segments that can be easily converted. This understanding will help provide insights to the marketing team for devising effective strategies to enhance customer engagement and drive sales.


## Step 1: Data Loading and Exploration
- **Objective:** Load the dataset and inspect its structure, handle missing values, and analyze descriptive statistics.

**Key Tasks:**
1. Load the dataset using `pandas`.
2. Check for missing values and overall dataset structure.
3. Generate summary statistics (mean, median, etc.) for numerical attributes.

This step helps us understand the dataset and prepare it for preprocessing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

# Step 1: Load and preprocess the data
data = pd.read_csv('Mall_Customers.csv')

# Display the first few rows of the dataset
print("First five rows of the dataset:")
print(data.head())

# Display dataset information
print("\nDataset Information:")
print(data.info())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Display descriptive statistics
print("\nDescriptive Statistics:")
print(data.describe())

# Display the unique values in the categorical column 'Genre'
print("\nUnique Values in 'Genre':")
print(data['Genre'].unique())
```

## Step 2: Data Preprocessing
- **Objective:** Prepare the data for clustering by encoding categorical variables and normalizing numerical features.

**Key Tasks:**
1. Encode categorical variables (e.g., Gender) using One-Hot Encoding.
   - This ensures categorical attributes are represented in numerical format suitable for clustering algorithms.
2. Normalize numerical variables (e.g., Age, Annual Income, Spending Score) using Min-Max Scaling.
   - Normalization ensures all attributes contribute equally to the clustering process.

This step ensures the dataset is clean and ready for clustering.

```python
# Use One-Hot Encoding for 'Genre'
one_hot_encoder = OneHotEncoder(drop='first', sparse_output=False)  # Corrected to 'sparse_output'
encoded_genre = one_hot_encoder.fit_transform(data[['Genre']])
encoded_genre_df = pd.DataFrame(encoded_genre, columns=one_hot_encoder.get_feature_names_out(['Genre']))

# Combine one-hot encoded genre with the original data
data = pd.concat([data, encoded_genre_df], axis=1)

# Normalize numerical attributes
scaler = MinMaxScaler()
data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(
    data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
)

# Define the feature matrix including one-hot encoded genre
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Genre_Male']]  # 'Genre_Male' is the one-hot encoded feature
```

## Step 3.1: KMeans Clustering
- **Objective:** Use the KMeans clustering algorithm to segment customers.

**Key Tasks:**
1. Apply the Elbow Method to determine the optimal number of clusters.
   - Plot Within-Cluster Sum of Squares (WCSS) vs. the number of clusters.
2. Fit the KMeans algorithm to the dataset with the chosen number of clusters.
3. Assign cluster labels to each data point.

The Elbow Method helps identify the best number of clusters, and the fitted KMeans model provides the cluster assignments.


```python
# Step 3.1: KMeans Clustering
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Optimal clusters based on the Elbow Method
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(X)
```

## Step 3.2: Hierarchical Clustering
- **Objective:** Use agglomerative hierarchical clustering to segment customers.

**Key Tasks:**
1. Compute the linkage matrix using different linkage methods (e.g., ward, complete, average).
2. Plot a dendrogram to visualize hierarchical relationships and determine the optimal number of clusters.
3. Assign cluster labels using the `fcluster` method.

Hierarchical clustering provides an alternative to KMeans, particularly useful when visualizing relationships between data points.


```python
# Step 3.2: Hierarchical Clustering
linked = linkage(X, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()

# Assign clusters using fcluster
data['Hierarchical_Cluster'] = fcluster(linked, t=optimal_clusters, criterion='maxclust')
```

## Step 3.3: DBSCAN Clustering
- **Objective:** Use the DBSCAN algorithm to cluster customers based on density.

**Key Tasks:**
1. Set parameters (`eps` and `min_samples`) to control the density threshold for forming clusters.
2. Fit the DBSCAN model to the dataset and assign cluster labels.
   - Noise points will be assigned the label `-1`.

DBSCAN is particularly effective for identifying clusters of varying shapes and handling noise points.

```python
# Step 3.3: DBSCAN Clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(X)

# Visualize DBSCAN Clustering Results
plt.figure(figsize=(8, 5))
plt.scatter(
    data['Annual Income (k$)'], 
    data['Spending Score (1-100)'], 
    c=data['DBSCAN_Cluster'], 
    cmap='viridis', 
    marker='o', 
    edgecolor='k'
)
plt.title("DBSCAN Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.colorbar(label='Cluster Label')
plt.show()
```

## Step 4: Cluster Evaluation
- **Objective:** Assess the quality of clustering using evaluation metrics.

**Key Metrics:**
1. **Davies-Bouldin Index (DBI):**
   - Measures the compactness and separation of clusters.
   - Lower values indicate better clustering.
2. **Dunn Index:**
   - Measures the ratio of minimum inter-cluster distance to maximum intra-cluster distance.
   - Higher values indicate better clustering.

Comparing DBI and Dunn Index across clustering techniques provides insights into which method performs best.

```python
# Step 4: Evaluate Clustering
def dunn_index(X, labels):
    unique_clusters = np.unique(labels)
    clusters = [X[labels == cluster] for cluster in unique_clusters if cluster != -1]
    intra_dists = [np.max(np.linalg.norm(cluster - cluster.mean(axis=0), axis=1)) for cluster in clusters]
    min_inter_dist = np.min([np.linalg.norm(c1.mean(axis=0) - c2.mean(axis=0))
                             for i, c1 in enumerate(clusters) for j, c2 in enumerate(clusters) if i < j])
    return min_inter_dist / max(intra_dists)

results = {}
for method, cluster_col in [('KMeans', 'KMeans_Cluster'), ('Hierarchical', 'Hierarchical_Cluster'), ('DBSCAN', 'DBSCAN_Cluster')]:
    labels = data[cluster_col].values
    dbi = davies_bouldin_score(X, labels) if len(np.unique(labels)) > 1 else np.inf
    dunn = dunn_index(X.values, labels)
    results[method] = {'DBI': dbi, 'Dunn Index': dunn}
```

## Step 5: Visualization
- **Objective:** Visualize the clusters generated by each technique.

**Key Tasks:**
1. Create 2D scatter plots for the clusters using `Annual Income` vs. `Spending Score`.
   - Use different colors to represent clusters.
   - Highlight noise points for DBSCAN.
2. Compare the visualizations to assess the cluster separability and patterns.

Visualizing clusters helps in interpreting results and understanding customer segments.

```python
# Step 5: Visualize Clusters
for method, cluster_col in [('KMeans', 'KMeans_Cluster'), ('Hierarchical', 'Hierarchical_Cluster'), ('DBSCAN', 'DBSCAN_Cluster')]:
    plt.figure(figsize=(8, 5))
    plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data[cluster_col], cmap='viridis', label=f'{method} Clusters')
    plt.title(f'{method} Clustering Visualization')
    plt.xlabel('Annual Income')
    plt.ylabel('Spending Score')
    plt.legend()
    plt.show()

# Step 7: Analyze Results
for method, metrics in results.items():
    print(f"{method} Clustering Results:")
    print(f"Davies-Bouldin Index: {metrics['DBI']}")
    print(f"Dunn Index: {metrics['Dunn Index']}")
```

## Conclusion: Clustering Techniques Comparison

From the clustering results obtained, we can draw the following insights about the performance of the different clustering techniques based on the **Davies-Bouldin Index (DBI)** and **Dunn Index**:

#### 1. **KMeans Clustering**
- **Davies-Bouldin Index:** 0.9954 (lowest among all techniques)
  - A lower DBI indicates better clustering, as it suggests less within-cluster variance and greater separation between clusters.
- **Dunn Index:** 0.7142
  - While not the highest, this index suggests moderate cluster compactness and separation.

**Inference:** KMeans performs the best overall, as it achieves the lowest DBI, indicating well-defined clusters with good separation and compactness.

#### 2. **Hierarchical Clustering**
- **Davies-Bouldin Index:** 1.0942
  - Slightly higher DBI compared to KMeans, suggesting that the clusters are less distinct than those in KMeans.
- **Dunn Index:** 0.7052
  - Almost similar to KMeans but slightly lower, indicating slightly poorer cluster separation.

**Inference:** Hierarchical clustering provides reasonable results but is slightly less effective compared to KMeans for this dataset.

#### 3. **DBSCAN Clustering**
- **Davies-Bouldin Index:** 1.8412 (highest among all techniques)
  - A high DBI indicates poor cluster quality, as the clusters have higher within-cluster variance and less separation.
- **Dunn Index:** 1.4257 (highest among all techniques)
  - The high Dunn Index suggests that DBSCAN identifies some well-separated clusters, though the overall cluster quality is inconsistent.

**Inference:** DBSCAN struggles with this dataset, likely due to the density parameters (`eps` and `min_samples`) not being optimal for the data distribution. However, it does identify well-separated clusters in specific areas of the data.

---

### Final Recommendation:
- **KMeans** is the best clustering technique for this dataset, as it achieves the most balanced and consistent results with a low Davies-Bouldin Index and reasonable Dunn Index.
- Hierarchical clustering is a valid alternative but slightly less effective.
- DBSCAN, while effective for some specific density-based problems, does not perform well for this dataset and requires further tuning to improve results.

### Business Implications:
Using KMeans clustering, we can segment customers into distinct groups effectively. These clusters can help target marketing strategies toward specific customer segments, such as high-spending or low-income groups, to improve customer retention and sales.
